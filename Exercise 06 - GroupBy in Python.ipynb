```python
####### Import used cars datasets ######
#####################################################

import pandas as pd
import numpy as np
df1 = pd.read_csv('citibike-tripdata.csv') # df1: contains citibike data
df2 = pd.read_csv('usedcars.csv') # df2: contains used car data

df2 = df2.replace("?", np.nan) 
df2=df2.dropna(subset=["price"], axis=0) # drop row that missing value in "price"
df2["price"] = df2["price"].astype("int") # convert "price" from object (string) to integer
df2.dtypes
df2.head(20)
```

```python
print(df2["body-style"].count())
print(df2["drive-wheels"].count())
df2["drive-wheels"].value_counts()
```

```python
df2=df2.dropna(subset=["body-style"], axis=0) # drop row that missing value in "body-style"
df2=df2.dropna(subset=["drive-wheels"], axis=0)
df2.shape
```

```python
df2["body-style"].count()
```

```python
# Assume we are interested in finding the average price of vehicles and observe how they
#-- differ between different types of body-style and drive wheels variables

df2_data = df2[["drive-wheels", "body-style", "price"]] # pick out the three data columns
#-- we are interested in

df2_group = df2_data.groupby(["drive-wheels","body-style"],as_index=False).mean() 
# as_index=False: indicate to groupby() that you do not want to set the column ID as index. Allows
#-- us to check the condition on a common column and not on an index
# group the reduced data 
#-- according to drive wheels and body style. Since we are intested in knowing how the 
#-- average price differs across the board, we take the mean of each group

df2_group # The data is now grouped into subcategories and only the average price of each
#-- subcategory is shown. We can see that rwd-convertible and rwd-hardtop has the highest value
#-- while 4wd-hatchbacks have the lowest value
```

```python
df2_data = df2[["drive-wheels", "body-style", "price"]]
df2_group = df2_data.groupby(["drive-wheels","body-style"],as_index=False).mean()
df2_group
### notice that there is no 4wd convertible and no 4wd hardtop
```

```python
# The above table is not easy to see/compare so we want to tranform it into Pivot table:
#-- 
df2_pivot = df2_group.pivot(index="drive-wheels", columns="body-style")
```

```python
df2_pivot
```

```python
pdf2_usedcar_pivot=pd.pivot_table(df2[["drive-wheels","body-style","price"]],index="drive-wheels",columns="body-style")
pdf2_usedcar_pivot
```
